```python
import numpy as np
from matplotlib import pyplot as plt


def motivation_dynamic(motivations):
    """Displays change of motivation
    state over time"""
    # course or date dependency
    x = np.linspace(0, 1, 11)
    y = motivations
    plt.title("Изменения мотивации со временем")
    plt.xlabel('Время пользования')
    plt.ylabel('Мотивация(амотивация)')
    ax = plt.gca()
    ax.set_ylim([-16, 48])
    plt.plot(x, y, 'o-g')
    plt.grid(True)
    # return plt
    plt.show()
```

```python

motivations = [13, 17, 24, 14, 5, -8, 45, None, None, 22, 34]

motivation_dynamic(motivations=motivations)
```
